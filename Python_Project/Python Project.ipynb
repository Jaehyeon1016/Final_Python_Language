In this project, I will display the graph and table that shows the COVID-19 confirmed cases in South Korea, and I will compare them to see the correlation between them.

The data on COVID-19 cases are from [KOSIS](https://kosis.kr/statHtml/statHtml.do)


```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
month_covid = pd.read_csv("./month.csv",
                          index_col=0, parse_dates=True)

month_covid.head()
```

The code brings the file that contains data and defines the first column of it. This table shows the COVID-19 newly confirmed cases in South Korea monthly.

```python
month_covid.plot.bar()
```

This graph shows that the newly confirmed cases were extremely increased in August and November. According to the News, in August, the corona virus started to spread in Seoul, which is the capital of South Korea, due to religion meetings, and in November, the corona virus spreaded again due to alleviated government policy.

```python
cumulative_covid = pd.read_csv("./released.csv",
                          index_col=0, parse_dates=True)

cumulative_covid.head()
```

The code does the same thing explained above. This table shows Cumulative total COVID-19 Confirmed cases and cumulative total people released from quaranine daily.

```python
cumulative_covid.plot()
```

```python
route_covid = pd.read_csv("./route.csv",
                          index_col=0, parse_dates=True)

route_covid.head()
```
